```python
top_2015 = affinio[affinio['subject']=='Affinio’s Best of 2015 - Our Year In Review']
```

```python
top_2015['content']
```

```python
top_2015['word_count']=gl.text_analytics.count_words(top_2015['content'])
```

```python
print top_2015['word_count']
```

```python
#now-our word_count is in dictionary---lets turn it into table
top_2015_table = top_2015[['word_count']].stack('word_count',new_column_name=['word','count'])
```

```python
top_2015_table.head()
```

```python
top_2015_table.sort('count',ascending=False)
```

```python
#lets now compute based on TF-IDF

affinio['word_count']=gl.text_analytics.count_words(affinio['content'])
```

```python
affinio.head()
```

```python
tfidf = gl.text_analytics.tf_idf(affinio['word_count'])
```

```python
affinio['tfidf'] = tfidf
```

```python
affinio.head()
```

```python
top_2015= affinio[affinio['subject']=='Affinio’s Best of 2015 - Our Year In Review']
```

```python
top_2015[['tfidf']].stack('tfidf', new_column_name=['word','tfidf']).sort('tfidf',ascending=False)
```

```python
knn_model = gl.nearest_neighbors.create(affinio, features=['tfidf'],label='subject')
```

```python
knn_model.query(top_2015)
```

```python
santa= affinio[affinio['subject'] == 'Santa Using Social? How Santa Uses Data to Find His Audience']
```

```python
knn_model.query(santa)
```
